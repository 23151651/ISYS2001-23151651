## Part 3: Building the Budget Tracker

Now for the real thing. Ask AI to help you write a program that:
- Takes the user's monthly income
- Takes up to 3 expenses
- Calculates what's left over
- Warns the user if the budget goes negative

> 💡 **Pro tip:** The more specific your prompt, the less clean-up you'll need to do afterwards. "Write a budget tracker" gets you something generic; describing the exact inputs, outputs, and edge cases gets you something closer to what you actually need.


```python
# Budget Calculator

# Ask the user for their monthly income
income = float(input("Enter your monthly income: $"))

# Ask for up to 3 expenses
expense1 = float(input("Enter expense 1: $"))
expense2 = float(input("Enter expense 2: $"))
expense3 = float(input("Enter expense 3: $"))

# Calculate total expenses and remaining budget
total_expenses = expense1 + expense2 + expense3
remaining = income - total_expenses

# Display the results
print("\n--- Budget Summary ---")
print(f"Monthly Income: ${income:.2f}")
print(f"Total Expenses: ${total_expenses:.2f}")
print(f"Money Left Over: ${remaining:.2f}")

# Warn the user if the budget is negative
if remaining < 0:
    print("Warning: Your budget is negative! You are spending more than you earn.")
else:
    print("Good job! Your budget is within your income.")
```

## Part 4: Enhancement Challenge

Pick ONE feature to add with AI's help:
- 🏷️ Expense categories (food, transport, entertainment)

- ✅ Input validation (no negative numbers)
- 📝 A detailed budget summary with advice


```python
# Budget Calculator with Percentage of Income Spent

# Ask the user for their monthly income
income = float(input("Enter your monthly income: $"))

# Ask for up to 3 expenses
expense1 = float(input("Enter expense 1: $"))
expense2 = float(input("Enter expense 2: $"))
expense3 = float(input("Enter expense 3: $"))

# Calculate total expenses and remaining budget
total_expenses = expense1 + expense2 + expense3
remaining = income - total_expenses

# Calculate the percentage of income spent
percentage_spent = (total_expenses / income) * 100

# Display the results
print("\n--- Budget Summary ---")
print(f"Monthly Income: ${income:.2f}")
print(f"Total Expenses: ${total_expenses:.2f}")
print(f"Money Left Over: ${remaining:.2f}")
print(f"Percentage of Income Spent: {percentage_spent:.1f}%")

# Warn the user if the budget is negative
if remaining < 0:
    print("Warning: Your budget is negative! You are spending more than you earn.")
else:
    print("Good job! Your budget is within your income.")
```
